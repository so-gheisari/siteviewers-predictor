```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, cross_val_predict, cross_validate
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale, normalize, minmax_scale
import math
from sklearn.metrics import mean_absolute_error
```

```python
input_rides = pd.read_csv('/Users/soulmaz/Input.csv')
rides_data = input_rides.drop(['Unnamed: 0','Year', 'Hour-from-Midnight'], axis=1)
selected_days = rides_data.iloc[0:11520,:]
```

```python
rides_data
```

```python
be = 0
en =1440
i = 1
for day in range(0,8):
    day_min = selected_days.iloc[be:en,4]
    day_rides = selected_days.iloc[be:en,5]
    plt.subplot(4, 2, i)
    plt.plot(day_min, day_rides)
    plt.title('rides of %d'%(i))
    i = i+1
    be = en
    en = en+1440
plt.show()
```

```python
x = rides_data.drop('Rides', axis=1)
y = rides_data.Rides
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
lin_reg_y_predict = lin_reg.predict(x_test)
import math
reg_mse = mean_squared_error(y_test,lin_reg_y_predict)
print(math.sqrt(reg_mse))
plt.scatter(y_test, lin_reg_y_predict)
plt.xlabel('rides')
plt.ylabel('predicted_rides')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=(3,6))
X_poly = poly_reg.fit_transform(x_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y_train)
lin_reg_2_y_predict = lin_reg_2.predict(poly_reg.fit_transform(x_test))
plt.plot(y_test, lin_reg_2_y_predict)
plt.show()
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
HGBR_reg = HistGradientBoostingRegressor(max_iter=10000)
HGBR_reg.fit(x_train,y_train)
HGBR_y_predict = HGBR_reg.predict(x_test)
```

```python
import math
from sklearn.metrics import mean_absolute_error
reg_mse = mean_squared_error(y_test,HGBR_y_predict)
print(math.sqrt(reg_mse))
print(mean_absolute_error(y_test,HGBR_y_predict))
plt.scatter(y_test, HGBR_y_predict)
plt.xlabel('rides')
plt.ylabel('predicted_rides')
plt.show()
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
reg1 = ExtraTreesRegressor(max_features=0.3)
reg2 = RandomForestRegressor(n_estimators=15, max_features=0.3)
reg3 = GradientBoostingRegressor(n_estimators=100, subsample=0.5, min_samples_leaf=25, max_features=1,random_state=42)
```

```python
final_estimator = HistGradientBoostingRegressor(max_iter=1000)
sreg = StackingRegressor(estimators=[('gb', reg1), ('rf', reg2), ('lr', reg3)],final_estimator=final_estimator)
sreg.fit(x_train, y_train)
sreg_y_predict = sreg.predict(x_test)
```

```python
sreg_mse = mean_squared_error(y_test,sreg_y_predict)
print(math.sqrt(sreg_mse))
print(mean_absolute_error(y_test, sreg_y_predict))
plt.scatter(y_test, sreg_y_predict)
plt.xlabel('rides')
plt.ylabel('predicted_rides')
plt.show()
```
